```python
sns.histplot(df['G3'],bins=10,kde=True)
plt.title("DISTRIBUTION OF FINAL GRADES ")
plt.show()
```

FROM THE ABOVE DISTRIBUTION WE CAN UNDERSTAND THAT G3 FROMS A CENTRALLY-SKEWED NORMAL DISTRIBUTION WITH AVG STUDENTS SCORING AROUND 10 AS DISPLAYED EARLIER

```python
sns.scatterplot(x=df['G2'] , y=df['G3'])
plt.title("RECENT PERFORMANCE VS OVERALL PERFORMANCE")
plt.show()
```

```python
sns.scatterplot(x=df['studytime'],y=df['G3'])
plt.title("RELATION OF FINAL RESULT WRT STUDYTIME")
plt.show()
```

```python
sns.scatterplot(x=df['absences'],y=df['G3'])
plt.title("RELATION OF FINAL RESULT WRT ABSENCE IN CLASS")
plt.show()
```

```python
sns.boxplot(x=df["sex"], y=df["G3"])
plt.title("Gender vs Final Performance")
plt.show()
```

```python
sns.scatterplot(x="studytime", y="G3", hue="sex", data=df)
plt.title("Study Time vs Final Grade (Grouped by Gender)")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
lb = [ 'G1' , 'G2' , 'G3' ]
g1=(df['G1']>df['G1'].mean()).sum()
g2=(df['G2']>df['G2'].mean()).sum()
g3=(df['G3']>df['G3'].mean()).sum()
plt.plot(lb,[g1,g2,g3])  
plt.title("Students Above Average in G1, G2, G3")
plt.xlabel("Exams")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.scatterplot(x=(df['G1']+df['G2'])/2, y=df["G3"])
plt.title("Average Previous Grade vs Final Grade")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
